### AN IMAGE TO PENCIL SKETCH WITH PYTHON.

STEPS:

1.Find an image to convert to a pencil sketch.

2.Read the image in RBG format and convert it to a grayscale image. This will turn an image into a classic black and white photo.

3.Invert the grayscale image (negative image), this will be our inverted grayscale image.

4.Create the pencil sketch by mixing the grayscale image with the inverted blurry image.

```python
import cv2
import matplotlib.pyplot as plt
     
```

```python
img = cv2.imread("Mahatma_Gandhi_photo.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(6,6))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")

plt.title("GrayScale Image")
plt.show()
```

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(6,6))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(6,6))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(6,6))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
